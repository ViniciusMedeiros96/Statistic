### Quantas ocorrências de cada categoria nós temos?

#### Distribuição de frequências absolutas

```python
file["Grau de Instrução"].value_counts()
```

* Número de ocorrências = frequência
* Frequência absoluta: contagem direta de ocorrências de cada categoria

#### Distribuição de frequências relativas

```python
file["Grau de Instrução"].value_counts(normalize=True)*100 #normalize=True me dará a porcentagem dos valores
```

* frequencia relativa: $f_r = \dfrac{f_a}{N}$
em que $f_a$ é a frequencia absoluta e $N$ é o número total de ocorrências

#### Ex:

```python
df2 = pd.DataFrame({"sexo": ['H', 'M', 'M', 'M', 'H', 'H', 'M', 'H', 'H', 'H'], 
                    "acidentes":[1, 5, 6, 3, 2, 5, 1, 2, 1, 4]})
df2
```
